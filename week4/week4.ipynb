```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('gold.csv')
data.head()
data
```

```python
x=np.array(data['Year'])
y=np.array(data['Price (24 karat per 10 grams)'])
x1=x.mean()
y1=y.mean()
s=np.size(x)
sl=np.sum(x*y)-s*x1*y1
ope=np.sum(x*x)-s*x1*x1
slope=sl/ope
slope
```

```python
intercept=y1-slope*x1
intercept
```

```python
ypred=intercept+slope*x
ypred
```

```python
plt.scatter(x, y, color = "b",marker = "o", s = 15)
plt.plot(x, ypred, color = "r")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#using scikit learn
a=np.array(data['Year']).reshape(-1,1)
b=np.array(data['Price (24 karat per 10 grams)'])
model=LinearRegression().fit(a,b)
print(model.intercept_)
print(model.coef_)
```

```python
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y,ypred))
mse=mean_squared_error(y,ypred)
print(rmse ,"\n", mse)
```

```python
est=intercept+slope*2025
print('Gold_2025:',est/10)
```
